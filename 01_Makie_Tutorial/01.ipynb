### <b> 3. Empezar a graficar

Como primer ejemplo vamos a graficar una función simple: 
$$f(x) = x^2$$
* Construir los vectores de $x$ y $f(x)$:

```julia
#Crear vector de x
x = -5:0.5:5
#Crear vector de f(x)
f = x.^2;
```

* Construir la grafica:

```julia
#Grafica simple:
lines(x,f)
```

```julia
#Grafica de puntos
scatter(x,f)
```

```julia
#Ambas graficas:
lines(x,f)
scatter!(x,f)
current_figure()
```
