### <b> Empezar a graficar </b>
Retomando la función:
$$f(x) = x^2$$

```julia
#Crear vector de x
x = -5:0.5:5
#Crear vector de f(x)
f = x .^ 2;
```

* Construir la gráfica como objeto:

```julia
# Grafica como un objeto:
fig = Figure()
# Crear ejes:
ax = Dict()
ax[1] = Axis(fig[1, 1])
fig
```

```julia
# Insertar las graficas de lineas y puntos
lines!(ax[1], x, f)
scatter!(ax[1], x, f)
fig
```

* Agregar nombre a los ejes:

```julia
# Crear figura:
fig = Figure(size=(800, 800))           # Modificar el tamaño de de la figura

# Modificar ax[1]
ax[1] = Axis(fig[1, 1];
    title=L"\textbf{Grafica de f(x)}",  # Nombre de la grafica
    xlabel=L"x",                        # Nombre del eje x
    ylabel=L"f(x)",                     # Nombre del eje y
    titlesize=30,                       # Tamaño del titulo
    xlabelsize=30,                      # Tamaño del eje x
    ylabelsize=30,                      # Tamaño del eje y 
)
fig
```

```julia
# Insertar las graficas de lineas y puntos
lines!(ax[1], x, f)
scatter!(ax[1], x, f)
fig
```

```julia
# Crear figura:
fig = Figure(size=(800, 600))           # Modificar el tamaño de de la figura

# Modificar ax[1]
ax[1] = Axis(fig[1, 1];
    title=L"\textbf{Grafica de f(x)}",  # Nombre de la grafica
    xlabel=L"x",                        # Nombre del eje x
    ylabel=L"f(x)",                     # Nombre del eje y
    titlesize=30,                       # Tamaño del titulo
    xlabelsize=30,                      # Tamaño del eje x
    ylabelsize=30,                      # Tamaño del eje y 
)
```

* Estilo de las gráficas:

```julia
# Insertar las graficas de lineas
lines!(ax[1], x, f;
    label=L"\text{Simulación}",     # Nombre de la linea
    color=:Red,                     # Color de la linea
    linestyle=:dash,                # Tipo de linea (:solid, :dot, :dash, :dashdot, :dashdotdot)
    linewidth=3                     # Tamaño de la linea
)
fig
```

```julia
# Insertar la grafica de puntos:
scatter!(ax[1], x, f;
label = L"\text{Datos Experimentales}",  # Nombre de la linea
marker=:circle,                          # Tipo de marcador (:circle, :rect, :diamond, :hexagon, :cross)
strokecolor=:Black,                      # Color del contorno
color=:transparent,                      # Color del marcador
markersize=20,                           # Tamaño del marcador
strokewidth=2,                           # Tamaño del contorno
)
fig
```

* Colocar Legendas:

```julia
axislegend(
    position = :rb,                  # Posicion (:rb, :cb, :lb, :rt, :ct, lt)
    orientation=:vertical,           # Orientación
    labelsize=20,                    # Tamaño de texto
    patchsize=(20, 0),               # Tamaño de los marcadore
    framevisible=false,              # Eliminar el cuadro
    )
fig
```
