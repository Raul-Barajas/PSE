## <b> <font color='Purple'> Parte 3: Configuración Avanzada de los Ejes

```julia
using CairoMakie
```

### <b> Empezar a graficar </b>
Retomando la función:
$$f(x) = x^2$$

```julia
#Crear vector de x
x = -5:0.5:5
#Crear vector de f(x)
f = x .^ 2;
```

* Configurar ejes:

```julia
# Crear figura:
fig = Figure(size=(800, 800))           # Modificar el tamaño de de la figura
ax = Dict()

# Modificar ax[1]
ax[1] = Axis(fig[1, 1];
    xticks=-6:1:6,                          # Rango del eje x
    yticks=0:2:30,                          # Rango del eje y
    limits=((-6, 6), (-1, 30)),             # Limites de la grafica
    title=L"\textbf{a) Grafica de f(x)}",   # Nombre de la grafica
    xlabel=L"x",                            # Nombre del eje x
    ylabel=L"f(x)",                         # Nombre del eje y
    titlealign=:left,                       # Posición del titulo (:left, :center, :rigth)
    yaxisposition=:right,                   # Posición del eje y (:left, :rigth)
    xaxisposition=:top,                     # Posición del eje x (:top, :bottom)
    titlesize=30,                           # Tamaño del titulo
    xlabelsize=35,                          # Tamaño del eje x
    ylabelsize=35,                          # Tamaño del eje y 
    xticklabelsize=25,                      # Tamaño del eje x (numeros)
    yticklabelsize=25,                      # Tamaño del eje y (numeros)
    ylabelcolor=:Red,                       # Color del eje y
    yticklabelcolor=:Blue,                  # Color del eje y (numeros)
    xlabelcolor=:Red,                       # Color del eje y
    xticklabelcolor=:Blue,                  # Color del eje y (numeros)
    xminorticksvisible=true,                # Lineas de fondo en x  
    yminorticksvisible=true,                # Lineas de fondo en y
    xminorgridvisible=true,                 # Lineas de fondo visibles x
    yminorgridvisible=true,                 # Lineas de fondo visibles y
)
fig
```

* Insertar las gráficas:

```julia
# Insertar las graficas de lineas
lines!(ax[1], x, f;
    label=L"\text{Simulación}",     # Nombre de la linea
    color=:Black,                   # Color de la linea
    linestyle=:dash,                # Tipo de linea (:solid, :dot, :dash, :dashdot, :dashdotdot)
    linewidth=3                     # Tamaño de la linea
)
fig
```

```julia
# Insertar la grafica de puntos:
scatter!(ax[1], x, f;
    label=L"\text{Datos Experimentales}",    # Nombre de la linea
    marker=:circle,                          # Tipo de marcador (:circle, :rect, :diamond, :hexagon, :cross)
    strokecolor=:Red,                        # Color del contorno
    color=:transparent,                      # Color del marcador
    markersize=20,                           # Tamaño del marcador
    strokewidth=2,                           # Tamaño del contorno
)
fig
```

* Colocar Legendas:

```julia
axislegend(
    position=:rt,                  # Posicion (:rb, :cb, :lb, :rt, :ct, lt)
    orientation=:vertical,           # Orientación
    labelsize=25,                    # Tamaño de texto
    patchsize=(20, 0),               # Tamaño de los marcadore
    framevisible=false,              # Eliminar el cuadro
)
fig
```
