```julia
using CairoMakie
using XLSX 
using DataFrames
using LaTeXStrings
```

### <b> 1. Importar Datos de Excel

```julia
# Leer los datos desde el archivo Excel
Data = XLSX.readxlsx("00.xlsx")
;
```

Guardar los datos:

```julia
# Crear diccionarios para almacenar los datos
x = Dict()
y = Dict()
name_x = Dict()
name_y = Dict()

# Almacenar los datos en los diccionarios
x[1] = Array{Float64}(Data["01"]["A2:A22"])
y[1] = Array{Float64}(Data["01"]["B2:B22"])

# Almacenar los nombres en los diccionarios
name_x[1] = Array{String}(Data["01"]["A1:A1"])
name_y[1] = Array{String}(Data["01"]["B1:B1"])
;
```

* Crear grafica:

```julia
# Crear figura:
fig = Figure(size=(800, 800))           # Modificar el tamaño de de la figura
ax = Dict()

# Modificar ax[1]
ax[1] = Axis(fig[1, 1];
    xticks=0:1:10,                                      # Rango del eje x
    yticks=-0:0.4:4,                                    # Rango del eje y
    #limits=((0, 10.5), (0, 4.5)),                       # Limites de la grafica
    #title=L"\textbf{d) Función Coseno Hiperbolico}",   # Nombre de la grafica
    xlabel=latexstring(name_x[1][1]),                   # Nombre del eje x
    ylabel=latexstring(name_y[1][1]),                   # Nombre del eje y
    titlealign=:left,                                   # Posición del titulo (:left, :center, :rigth)
    yaxisposition=:left,                                # Posición del eje y (:left, :rigth)
    xaxisposition=:bottom,                              # Posición del eje x (:top, :bottom)
    titlesize=30,                                       # Tamaño del titulo
    xlabelsize=30,                                      # Tamaño del eje x
    ylabelsize=30,                                      # Tamaño del eje y 
    xticklabelsize=20,                                  # Tamaño del eje x (numeros)
    yticklabelsize=20,                                  # Tamaño del eje y (numeros)
    ylabelcolor=:Black,                                 # Color del eje y
    yticklabelcolor=:Black,                             # Color del eje y (numeros)
    xlabelcolor=:Black,                                 # Color del eje y
    xticklabelcolor=:Black,                             # Color del eje y (numeros)
    xminorticksvisible=true,                            # Lineas de fondo en x  
    yminorticksvisible=true,                            # Lineas de fondo en y
    xminorgridvisible=true,                             # Lineas de fondo visibles x
    yminorgridvisible=true,                             # Lineas de fondo visibles y
)
```

```julia
scatter!(ax[1], x[1][:], y[1][:];
label = L"\text{Datos Experimentales}",  # Nombre de la linea
marker=:circle,                          # Tipo de marcador (:circle, :rect, :diamond, :hexagon, :cross)
strokecolor=:Red,                        # Color del contorno
color=:transparent,                      # Color del marcador
markersize=20,                           # Tamaño del marcador
strokewidth=2,                           # Tamaño del contorno
)
fig
```

```julia
# Insertar las graficas de lineas
X = 0:0.01:10
Y = (X).^0.5
lines!(ax[1], X, Y;
    label=L"\text{Simulación}",     # Nombre de la linea
    color=:Black,                   # Color de la linea
    linestyle=:dash,                # Tipo de linea (:solid, :dot, :dash, :dashdot, :dashdotdot)
    linewidth=3                     # Tamaño de la linea
)
fig
```

```julia
axislegend(
    position=:lt,                  # Posicion (:rb, :cb, :lb, :rt, :ct, lt)
    orientation=:vertical,           # Orientación
    labelsize=25,                    # Tamaño de texto
    patchsize=(20, 0),               # Tamaño de los marcadore
    framevisible=false,              # Eliminar el cuadro
)
fig
```
